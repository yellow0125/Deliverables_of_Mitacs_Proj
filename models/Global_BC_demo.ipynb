```python
# ==================== MMSeg → Spotlight Visualizer (Images OR Video) ====================
# - MODE = "images": process a folder of images → per-image overlays
# - MODE = "video":  process a video → side-by-side video; optional per-frame PNGs
#
# - Overlay shows NO color below MIN_VIS_RISK
# - For >= MIN_VIS_RISK, colors are Yellow→Red (no greens/blues)
# - Colorbar matches overlay and shows a midpoint notch; left half is dimmed/hidden
#
# Colab-friendly. Works with MMSeg 1.x (mmengine) and 0.x.
# ---------------------------------------------------------------------------------------

import os, cv2, sys, re, numpy as np, traceback, glob
from pathlib import Path

# -------------------- Mode & Paths --------------------
MODE = "video"   # "images" or "video"

# If MODE=="images"
INPUT_FRAMES_DIR  = "/content/frames"          # folder with source frames
OUTPUT_FRAMES_DIR = "/content/output_frames"   # folder for processed overlays

# If MODE=="video"
INPUT_VIDEO  = "/content/input.mp4"
OUTPUT_VIDEO = "/content/output_spotlight.mp4"
SAVE_PNGS    = False                           # also save processed frames from video to PNGs
SAVE_DIR     = "/content/spotlight_frames"
FNAME_PREFIX = "frame"

# -------------------- Model --------------------
CONFIG_PATH  = "/content/config_output.py"
CKPT_PATH    = "/content/mmsegmentation/iter_2700.pth"

# -------------------- Controls --------------------
TARGET_FPS   = 0      # for video; None or 0 → keep all frames
DURATION_S   = 0     # for video; 0 → full video
MAX_W        = 960     # downscale for speed; None to keep original
SIDE_BY_SIDE = True    # for video: Original | Processed
TITLE_FONT_SCALE = 0.8
BORDER_DIVIDER   = 4
DEBUG        = True

# ---- Risk & shaping ----
RISK_WEIGHTS = {
    "beetle-fire": 1.0,
    "dead-tree":   0.7,
    "debris":   0.7,
    "null":        0.0,
}
ENABLE_PERCENTILE_NORM = True
PERC_LO, PERC_HI       = 15, 92     # slightly more aggressive spread upward
SIGMOID_K, SIGMOID_M   = 10.0, 0.33 # push reds a bit earlier than 0.40
SMOOTH_RISK_SIGMA      = 2.0        # slight smoothing for stability

# ---- Background treatment (outside risk) ----
BG_DESAT_STRENGTH  = 0.55
BG_DIM_STRENGTH    = 0.22
BG_BLUR_SIGMA      = 0        # px (0 disables)
DILATE_PX          = 5

# ---- Glow overlay ----
HEATMAP_CMAP       = "CUSTOM"  # CUSTOM / JET / TURBO / INFERNO / MAGMA / VIRIDIS
GLOW_ALPHA_MAX     = 0.75
GLOW_GAMMA         = 0.70
GLOW_SOFTEN_SIGMA  = 2.0
MIN_VIS_RISK       = 0.08      # small lift so more of the gradient shows

# ---- Temporal smoothing (EMA on risk) [video only] ----
TEMPORAL_EMA       = True
EMA_ALPHA          = 0.7

# -------------------- MMSeg imports & version switch --------------------
try:
    from mmseg.apis import init_model, inference_model
    _MMSEG_V1 = True
except Exception:
    from mmseg.apis import init_model, inference_segmentor as inference_model
    _MMSEG_V1 = False

# -------------------- Utils --------------------
def ensure_bgr(img):
    if img is None: return None
    if img.ndim == 2: return cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    if img.shape[2] == 3: return img
    if img.shape[2] == 4: return cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
    return img[..., :3]

def valid_frame(img):
    return (img is not None) and hasattr(img, "shape") and img.size > 0

def downscale_keep_ratio(img, max_w):
    if not max_w or not valid_frame(img): return img
    h, w = img.shape[:2]
    if w <= max_w: return img
    nw = int(max_w); nh = max(1, int(h * (nw / w)))
    return cv2.resize(img, (nw, nh), interpolation=cv2.INTER_LINEAR)

def put_label(img, text, org, scale=0.8, color=(255,255,255), thickness=2):
    cv2.putText(img, text, org, cv2.FONT_HERSHEY_SIMPLEX, scale, (0,0,0), thickness+3, cv2.LINE_AA)
    cv2.putText(img, text, org, cv2.FONT_HERSHEY_SIMPLEX, scale, color, thickness, cv2.LINE_AA)

def compose_side_by_side(l_frame, r_frame):
    lh, lw = l_frame.shape[:2]
    rh, rw = r_frame.shape[:2]
    if lh != rh:
        r_frame = cv2.resize(r_frame, (rw, lh), interpolation=cv2.INTER_LINEAR)
    divider = np.full((lh, BORDER_DIVIDER, 3), 32, np.uint8)
    return np.concatenate([l_frame, divider, r_frame], axis=1)

def _cmap_code(name: str):
    name = (name or "").upper()
    table = {
        "JET": cv2.COLORMAP_JET, "TURBO": cv2.COLORMAP_TURBO,
        "INFERNO": cv2.COLORMAP_INFERNO, "MAGMA": cv2.COLORMAP_MAGMA, "VIRIDIS": cv2.COLORMAP_VIRIDIS
    }
    return table.get(name, cv2.COLORMAP_TURBO)

# -------------------- MMSeg helpers --------------------
def coerce_hw_mask(arr):
    if arr is None: return None
    if hasattr(arr, "detach"):  # torch tensor
        arr = arr.detach().cpu().numpy()
    arr = np.array(arr); arr = np.squeeze(arr)
    if arr.ndim == 3:
        if arr.shape[-1] < 8 and arr.dtype != np.int32: arr = np.argmax(arr, axis=-1)
        else: arr = arr[..., 0]
    if arr.ndim != 2: return None
    return arr.astype(np.int32, copy=False)

def mmseg_palette(model):
    try:
        meta = getattr(model, "dataset_meta", None)
        if meta and "palette" in meta and meta["palette"] is not None:
            pal_rgb = np.array(meta["palette"], dtype=np.uint8)
            return pal_rgb[:, ::-1]  # to BGR
    except Exception:
        pass
    # fallback
    n = 4
    try:
        if hasattr(model, "dataset_meta") and "classes" in model.dataset_meta and model.dataset_meta["classes"] is not None:
            n = max(1, len(model.dataset_meta["classes"]))
    except Exception:
        pass
    rng = np.random.default_rng(7)
    cols = rng.integers(32, 255, size=(n, 3), dtype=np.uint8)
    return cols[:, ::-1]  # to BGR

def mmseg_classes(model, palette_len):
    names = None
    try:
        if hasattr(model, "dataset_meta") and "classes" in model.dataset_meta and model.dataset_meta["classes"]:
            names = list(model.dataset_meta["classes"])
    except Exception:
        names = None
    if names is None:
        names = [f"Class_{i}" for i in range(palette_len)]
    if len(names) < palette_len:
        names += [f"Class_{i}" for i in range(len(names), palette_len)]
    elif len(names) > palette_len:
        names = names[:palette_len]
    return names

def build_id2w(class_names, risk_weights):
    pairs = []; mn = mx = None
    for k, v in (risk_weights or {}).items():
        try: val = float(v)
        except: continue
        pairs.append((k, val))
        mn = val if mn is None else min(mn, val)
        mx = val if mx is None else max(mx, val)
    def norm(x):
        if mn is None or mx is None or abs(mx - mn) < 1e-6: return float(np.clip(x,0,1))
        return float((x - mn) / (mx - mn))
    canon = []
    for k, v in pairs:
        if isinstance(k, str) and k.isdigit(): canon.append((int(k), norm(v)))
        else:
            tks = "".join(ch.lower() if ch.isalnum() else " " for ch in str(k)).split()
            canon.append((tks, norm(v)))
    id2w = np.zeros(len(class_names), np.float32)
    for cid, cname in enumerate(class_names):
        cname_tokens = "".join(ch.lower() if ch.isalnum() else " " for ch in str(cname)).split()
        w = 0.0
        for k, v in canon:
            if isinstance(k, int) and k == cid: w = v; break
        if w == 0.0:
            for k, v in canon:
                if isinstance(k, list) and any(tok in cname_tokens for tok in k): w = v; break
        id2w[cid] = w
    return np.clip(id2w, 0.0, 1.0)

def mmseg_predict(model, frame_bgr):
    try:
        pred = inference_model(model, frame_bgr)
    except Exception as e:
        print(f"[ERROR] inference failed: {e}")
        traceback.print_exc(limit=2)
        return None
    if _MMSEG_V1:
        if hasattr(pred, "pred_sem_seg") and hasattr(pred.pred_sem_seg, "data"):
            return coerce_hw_mask(pred.pred_sem_seg.data)
        if isinstance(pred, (list, tuple)) and len(pred) > 0:
            p0 = pred[0]
            if hasattr(p0, "pred_sem_seg") and hasattr(p0.pred_sem_seg, "data"):
                return coerce_hw_mask(p0.pred_sem_seg.data)
    if isinstance(pred, (list, tuple)) and len(pred) > 0:
        return coerce_hw_mask(pred[0])
    return coerce_hw_mask(pred)

# -------------------- Spotlight helpers --------------------
def normalize_percentile(risk, hi=97, lo=10):
    lo_v, hi_v = np.percentile(risk, lo), np.percentile(risk, hi)
    if hi_v - lo_v < 1e-6: return np.clip(risk, 0, 1)
    r = (risk - lo_v) / (hi_v - lo_v)
    return np.clip(r, 0, 1)

def risk_from_mask(mask, id2w):
    max_id = int(mask.max())
    if max_id + 1 > len(id2w):
        pad = np.zeros((max_id + 1 - len(id2w),), np.float32)
        id2w = np.concatenate([id2w, pad], axis=0)
    risk = id2w[mask].astype(np.float32)
    if SMOOTH_RISK_SIGMA and SMOOTH_RISK_SIGMA > 0:
        k = max(1, int(SMOOTH_RISK_SIGMA*3)*2+1)
        risk = cv2.GaussianBlur(risk, (k, k), SMOOTH_RISK_SIGMA)
    if ENABLE_PERCENTILE_NORM:
        risk = normalize_percentile(risk, hi=PERC_HI, lo=PERC_LO)
    # Sigmoid shaping
    if SIGMOID_K != 0:
        risk = 1.0 / (1.0 + np.exp(-SIGMOID_K * (risk - SIGMOID_M)))
    return np.clip(risk, 0, 1)

def background_treat(img_bgr, keep_mask_bool):
    base2 = img_bgr.copy().astype(np.float32)
    drop = ~keep_mask_bool

    # Desaturate
    hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV).astype(np.float32)
    s = hsv[...,1] / 255.0
    s[drop] = s[drop] * (1.0 - BG_DESAT_STRENGTH)
    hsv[...,1] = np.clip(s * 255.0, 0, 255)
    base2 = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2BGR).astype(np.float32)

    # Dim
    base2[drop] *= (1.0 - BG_DIM_STRENGTH)

    # Blur outside
    if BG_BLUR_SIGMA and BG_BLUR_SIGMA > 0:
        k = max(1, int(BG_BLUR_SIGMA*3)*2+1)
        blurred = cv2.GaussianBlur(base2.astype(np.uint8), (k, k), BG_BLUR_SIGMA).astype(np.float32)
        base2[drop] = blurred[drop]

    return np.clip(base2, 0, 255).astype(np.uint8)

# -------------------- Colormap utilities --------------------
def custom_bgr_heatmap_yellow_red_from01(r01):
    """
    Map 0..1 to BGR strictly Yellow(0,255,255) → Red(0,0,255).
    Used for 'visible' range (≥ MIN_VIS_RISK) after re-scaling to 0..1.
    """
    r = np.clip(r01, 0, 1).astype(np.float32)
    B = np.zeros_like(r, dtype=np.float32)
    G = 255.0 * (1.0 - r)     # 255 -> 0
    R = 255.0 * np.ones_like(r, dtype=np.float32)
    return np.stack([B, G, R], axis=-1).astype(np.uint8)

# -------------------- Colorbar (matches overlay) --------------------
def draw_colorbar(img, cmap_name="CUSTOM", width=280, height=18, label="Risk Level",
                  x=10, y=None, threshold=0.5, notch=False):
    """
    Yellow→Red colorbar with 'Low' and 'High' above the bar,
    and 'Risk Level' centered below the bar (same style/color as Low/High).
    Auto-adjusts position to avoid cropping.
    """
    out = img.copy()
    H, W = out.shape[:2]

    # ---------- sizes & gaps ----------
    fs_small = 0.45         # Low/High size
    fs_label = 0.6          # Risk Level size
    label_color = (255,255,255)

    # Compute Risk Level text height to reserve space below the bar
    label_text = label if label else ""
    tsize = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, fs_label, 1)[0]
    title_gap = (tsize[1] + 10) if label else 14   # vertical gap below bar for title

    # ---------- place bar (ensure it fits) ----------
    if y is None:
        y = H - height - (title_gap + 6)  # default near bottom with safe gap

    # If the label would go off-canvas, move bar up
    if y + height + title_gap > H:
        y = max(0, H - height - title_gap)

    # Also clamp X/Y inside left/right/top bounds
        # Also clamp X/Y inside left/right/top bounds
    x = int(max(0, min(x, max(0, W - width))))
    y = int(max(0, min(y, max(0, H - height - title_gap))))


    # Early out if bar wouldn't fit (very small frames)
    if x + width > W or y + height > H:
        return out

    # ---------- draw full Yellow→Red bar (no hidden grey) ----------
    if cmap_name.upper() == "CUSTOM":
        vals01 = np.linspace(0.0, 1.0, width, dtype=np.float32)[None, :]
        bar_full = custom_bgr_heatmap_yellow_red_from01(vals01.repeat(height, axis=0))
        out[y:y+height, x:x+width] = bar_full
    else:
        vals255 = np.linspace(0, 255, width, dtype=np.uint8)[None, :]
        bar = np.repeat(vals255, height, axis=0)
        bar = cv2.applyColorMap(bar, _cmap_code(cmap_name))
        out[y:y+height, x:x+width] = bar

    # (Notch intentionally ignored; keep notch=False in calls)

    # ---------- labels ----------
    # Low / High ABOVE the bar (same style as your processed text via put_label)
    tick_y = y - 8
    # If too close to top, nudge bar down a bit to keep labels visible
    if tick_y < 8:
        dy = 8 - tick_y
        y = min(y + dy, max(0, H - height - title_gap))
        tick_y = y - 8
        # redraw bar at new y
        if cmap_name.upper() == "CUSTOM":
            out[y:y+height, x:x+width] = bar_full
        else:
            out[y:y+height, x:x+width] = bar

    put_label(out, "Low",  (x, tick_y),               scale=fs_small, color=label_color, thickness=1)
    right_text = "High"
    right_size = cv2.getTextSize(right_text, cv2.FONT_HERSHEY_SIMPLEX, fs_small, 1)[0]
    right_x = x + width - right_size[0]
    put_label(out, right_text, (right_x, tick_y),      scale=fs_small, color=label_color, thickness=1)

    # Risk Level BELOW the bar (centered) — same style/color as Low/High
    if label:
        text_x = x + (width - tsize[0]) // 2
        text_y = y + height + (tsize[1] + 8)  # sits inside reserved title_gap
        # If still too low (tiny frames), clamp upward
        if text_y > H - 2:
            text_y = H - 2
        put_label(out, label, (text_x, text_y), scale=fs_label, color=label_color, thickness=1)

    return out




# -------------------- Glow with threshold --------------------
def apply_glow(base_bgr, risk01, alpha_max=0.75, gamma=0.70, cmap="CUSTOM"):
    """
    - No color below MIN_VIS_RISK (alpha=0).
    - For risk >= MIN_VIS_RISK, re-map to 0..1 and color:
        CUSTOM -> Yellow→Red (no greens/blues)
        OpenCV cmaps -> applied on the re-mapped 0..1
    """
    thr = float(MIN_VIS_RISK)
    r = np.clip(risk01, 0, 1).astype(np.float32)

    # mask for visible region
    vis_mask = (r >= thr)

    # re-map visible risk to 0..1 so 0==thr and 1==1
    r_vis = np.zeros_like(r, dtype=np.float32)
    if thr < 1.0:
        r_vis[vis_mask] = (r[vis_mask] - thr) / (1.0 - thr)  # 0..1

    # build heatmap
    if cmap.upper() == "CUSTOM":
        # continuous Yellow→Red using r_vis (0..1)
        r_cont = np.zeros_like(r, dtype=np.float32)
        r_cont[vis_mask] = r_vis[vis_mask]
        B = np.zeros_like(r_cont, dtype=np.float32)
        G = 255.0 * (1.0 - r_cont)
        R = 255.0 * np.ones_like(r_cont)
        hm = np.stack([B, G, R], axis=-1).astype(np.uint8)
    else:
        hm = cv2.applyColorMap((r_vis * 255).astype(np.uint8), _cmap_code(cmap))

    # optional softening
    if GLOW_SOFTEN_SIGMA and GLOW_SOFTEN_SIGMA > 0:
        k = max(1, int(GLOW_SOFTEN_SIGMA*3)*2+1)
        hm = cv2.GaussianBlur(hm, (k, k), GLOW_SOFTEN_SIGMA)

    # alpha only for visible region (raise with gamma)
    alpha_map = np.zeros_like(r, dtype=np.float32)
    alpha_map[vis_mask] = (np.power(r_vis[vis_mask], gamma)) * alpha_max

    # blend
    out = (base_bgr.astype(np.float32) * (1.0 - alpha_map[..., None]) +
           hm.astype(np.float32)       * (alpha_map[..., None]))
    out = np.clip(out, 0, 255).astype(np.uint8)

    # colorbar that matches the overlay policy (with notch & shaded left)
    out = draw_colorbar(out, cmap_name=HEATMAP_CMAP, label="Risk Level",
                    threshold=MIN_VIS_RISK, notch=False)

    return out

# -------------------- Init model --------------------
if not Path(CONFIG_PATH).exists():  print(f"[WARN] CONFIG_PATH not found: {CONFIG_PATH}")
if not Path(CKPT_PATH).exists():    print(f"[WARN] CKPT_PATH not found: {CKPT_PATH}")
print("Loading MMSeg model…")
model = init_model(CONFIG_PATH, CKPT_PATH, device="cuda" if cv2.cuda.getCudaEnabledDeviceCount()>0 else "cpu")
PALETTE = mmseg_palette(model)
CLASS_NAMES = mmseg_classes(model, len(PALETTE))
ID2W = build_id2w(CLASS_NAMES, RISK_WEIGHTS)
if DEBUG:
    nz = {CLASS_NAMES[i]: float(ID2W[i]) for i in range(len(CLASS_NAMES)) if ID2W[i] > 0}
    print("[Non-zero weights]", nz if nz else "{}")

# Identify background IDs
BG_IDS = {i for i, n in enumerate(CLASS_NAMES) if str(n).lower() in {"null","background","bg"}}
if not BG_IDS: BG_IDS = {0}

# -------------------- Core process (one frame) --------------------
def process_one(fr_bgr, risk_ema=None):
    pred = mmseg_predict(model, fr_bgr)
    if pred is None or pred.ndim != 2 or pred.size == 0:
        return fr_bgr.copy(), None
    H, W = fr_bgr.shape[:2]
    if pred.shape[:2] != (H, W):
        pred = cv2.resize(pred, (W, H), interpolation=cv2.INTER_NEAREST)

    risk = risk_from_mask(pred, ID2W)
    if risk_ema is not None:
        risk = EMA_ALPHA * risk_ema + (1 - EMA_ALPHA) * risk

    # Build soft keep mask for BG treatment
    non_bg = np.zeros_like(pred, np.uint8)
    for bid in BG_IDS:
        non_bg |= (pred != int(bid)).astype(np.uint8)
    if DILATE_PX > 0:
        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (DILATE_PX, DILATE_PX))
        keep_bool = cv2.dilate(non_bg, kernel, iterations=1).astype(bool)
    else:
        keep_bool = non_bg.astype(bool)

    base_treated = background_treat(fr_bgr, keep_bool)
    vis = apply_glow(base_treated, risk, alpha_max=GLOW_ALPHA_MAX, gamma=GLOW_GAMMA, cmap=HEATMAP_CMAP)

    if DEBUG:
        q = np.quantile(risk, [0.1, 0.25, 0.5, 0.75, 0.9])
        print(f"risk q10..q90: {q}")

    return vis, risk

# -------------------- Run: Images mode --------------------
def run_images():
    Path(OUTPUT_FRAMES_DIR).mkdir(parents=True, exist_ok=True)
    exts = ("*.png","*.jpg","*.jpeg","*.bmp","*.tif","*.tiff","*.webp")
    files = []
    for e in exts:
        files += glob.glob(str(Path(INPUT_FRAMES_DIR) / e))
    files = sorted(files, key=lambda p: (Path(p).stem, p))
    if len(files) == 0:
        print(f"[ERROR] No images found in {INPUT_FRAMES_DIR}")
        return

    print(f"[INFO] Processing {len(files)} images from {INPUT_FRAMES_DIR} → {OUTPUT_FRAMES_DIR}")
    for i, p in enumerate(files, 1):
        img = cv2.imread(p, cv2.IMREAD_COLOR)
        if not valid_frame(img):
            print(f"[WARN] Could not read {p}")
            continue
        img = downscale_keep_ratio(ensure_bgr(img), MAX_W)
        vis, _ = process_one(img, risk_ema=None)  # no temporal EMA for static images
        outp = Path(OUTPUT_FRAMES_DIR) / (Path(p).stem + "_overlay.png")
        cv2.imwrite(str(outp), vis)
        if DEBUG and (i % 10 == 0):
            print(f"  saved {i}/{len(files)}: {outp}")
    print("[OK] Image processing complete.")

# -------------------- Run: Video mode --------------------
def run_video():
    cap = cv2.VideoCapture(INPUT_VIDEO)
    if not cap.isOpened():
        print(f"[ERROR] Could not open video: {INPUT_VIDEO}")
        return

    src_fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) or 0)
    if TARGET_FPS and TARGET_FPS > 0:
        step = max(1, int(round(src_fps / TARGET_FPS)))
        out_fps = TARGET_FPS
    else:
        step = 1
        out_fps = src_fps

    max_frames = total_frames if not DURATION_S or DURATION_S <= 0 else int(min(total_frames, DURATION_S * src_fps))
    print(f"Source FPS: {src_fps:.2f} | step={step} | output FPS={out_fps:.2f} | frames≈{max_frames//step}")

    # Find first frame
    probe_ok, probe_frame = False, None
    for idx in range(0, max_frames, step):
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, fr = cap.read()
        if ret and valid_frame(fr):
            fr_rs = downscale_keep_ratio(ensure_bgr(fr), MAX_W)
            probe_ok, probe_frame = True, fr_rs
            break
    if not probe_ok:
        print("[ERROR] No readable frames at selected sampling.")
        return

    vis_probe, _ = process_one(probe_frame)
    left_probe  = probe_frame.copy()
    put_label(left_probe, "Original", (12, 28), scale=TITLE_FONT_SCALE)
    put_label(vis_probe, "Processed", (12, 28), scale=TITLE_FONT_SCALE)

    if SIDE_BY_SIDE:
        oh, ow = left_probe.shape[:2]
        rh, rw = vis_probe.shape[:2]
        if oh != rh:
            vis_probe = cv2.resize(vis_probe, (rw, oh), interpolation=cv2.INTER_LINEAR)
        out_size = (ow + BORDER_DIVIDER + rw, oh)
    else:
        out_size = (vis_probe.shape[1], vis_probe.shape[0])

    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    writer = cv2.VideoWriter(OUTPUT_VIDEO, fourcc, float(out_fps), out_size)
    if not writer.isOpened():
        print("mp4v failed, trying XVID…")
        fourcc = cv2.VideoWriter_fourcc(*"XVID")
        writer = cv2.VideoWriter(OUTPUT_VIDEO.replace(".mp4", ".avi"), fourcc, float(out_fps), out_size)
    if not writer.isOpened():
        print("[ERROR] Could not open any video writer.")
        return

    if SAVE_PNGS:
        Path(SAVE_DIR).mkdir(parents=True, exist_ok=True)

    risk_ema = None
    frame_count = 0
    saved_idx = 0

    for idx in range(0, max_frames, step):
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, fr = cap.read()
        if not ret or not valid_frame(fr):
            if DEBUG: print(f"[WARN] read fail at frame {idx}")
            continue

        fr_rs = downscale_keep_ratio(ensure_bgr(fr), MAX_W)
        left  = fr_rs.copy(); put_label(left, "Original", (12, 28), scale=TITLE_FONT_SCALE)

        vis, risk_now = process_one(fr_rs, risk_ema if (TEMPORAL_EMA and risk_ema is not None) else None)
        if risk_now is not None and TEMPORAL_EMA:
            risk_ema = risk_now if risk_ema is None else (EMA_ALPHA * risk_ema + (1 - EMA_ALPHA) * risk_now)

        put_label(vis, "Processed", (12, 28), scale=TITLE_FONT_SCALE)
        if SIDE_BY_SIDE:
            lh, lw = left.shape[:2]
            rh, rw = vis.shape[:2]
            if lh != rh:
                vis = cv2.resize(vis, (rw, lh), interpolation=cv2.INTER_LINEAR)
            frame_out = np.concatenate([left, np.full((lh, BORDER_DIVIDER, 3), 32, np.uint8), vis], axis=1)
        else:
            frame_out = vis

        writer.write(frame_out)
        frame_count += 1

        if SAVE_PNGS:
            saved_idx += 1
            cv2.imwrite(str(Path(SAVE_DIR) / f"{FNAME_PREFIX}_{saved_idx:04d}_overlay.png"), vis)

        if DEBUG and (frame_count % 10 == 0):
            print(f"Processed {frame_count} frames…")

    cap.release()
    writer.release()
    print(f"[OK] Wrote {frame_count} frames to {OUTPUT_VIDEO}")
    if SAVE_PNGS:
        print(f"[OK] Saved per-frame overlays to {SAVE_DIR}")

# -------------------- Entrypoint --------------------
if MODE.lower() == "images":
    run_images()
elif MODE.lower() == "video":
    run_video()
else:
    print(f"[ERROR] Unknown MODE: {MODE}. Use 'images' or 'video'.")
```
